# Customer Segmentation Using K-Means clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("C:/Users/Admin/Desktop/Kaveri/Prodigyinfo/Mall_Customers.csv")
```

```python
data
```

```python
X = data.iloc[:, 3:].values
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
data['Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(10, 7))
```

```python
for i in range(5):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s=100, label=f'Cluster {i+1}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of customers based on purchase history')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
